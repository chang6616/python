```python
import json
```

```python
file = open("程式語言HW2-登革熱.json" , encoding="utf8")
data = json.load(file)
type(data)
```

```python
dicdata = {"案例":data}
type(dicdata)
```

```python
case = dicdata.get("案例")
case
```

```python
district ="台北市"
count = 0

for i in case:
    if district == i.get("縣市"):
        count = count+1

print("在台北市發生的登革熱有", count, "筆")
```
